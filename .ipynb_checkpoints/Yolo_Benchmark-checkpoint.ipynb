```python
"""
Auteur : Adrien Ducourthial
Date   : 16/11/2019

Ce code python a pour objectif de faire un benchmark  de yolo entrainÃ© sur COCO 
"""
import numpy as np
import time
import cv2
import os

def initialisation_COCO():
    """
    Load coco's labels and add color for each of them
    """
    labelsPath = str(os.path.sep)+ "\yolo\coco.names"
    global LABELS
    LABELS = open("yolo\coco.names").read().strip().split("\n")
    np.random.seed(42)
    COLORS = np.random.randint(0, 255, size=(len(LABELS), 3),dtype="uint8")

def initialisation_Yolo(ver):
    """
    Load weights and cfg
    Create Net from Darknet
    """
    weightsPath = "yolo"+str(ver)+".weights"
    configPath = "yolo"+str(ver)+".cfg"
    global net 
    net = cv2.dnn.readNetFromDarknet(configPath, weightsPath)

def import_img(folder,names):
    """
    Load images from a specific folder.
    Input
        arg : folder path, files names
    Return
        list : list of pixels values 
    """
    return [cv2.imread(str(folder)+str(i)) for i in names]

        
def run_YOLO(images):
    imgname = 0
    outlist =[]
    ln = net.getLayerNames()
    ln = [ln[i[0] - 1] for i in net.getUnconnectedOutLayers()]
    for image in images:
        (W,H) = image.shape[:2]
        boxes = []
        classIDs = []
        confidences=[]
        det = []
        imgname +=1
        
        blob = cv2.dnn.blobFromImage(image, 1 / 255.0, (416, 416),swapRB=True, crop=False)
        net.setInput(blob)
        start = time.time()
        layerOutputs = net.forward(ln)
        end = time.time()
        for output in layerOutputs:
            for detection in output:
                scores = detection[5:]
                classID = np.argmax(scores)
                confidence = scores[classID]
                if confidence > 0.5:
                    temp = []
                    box = detection[0:4] * np.array([H, W, H, W])
                    (centerX, centerY, width, height) = box.astype("int")
                    #print((centerX, centerY, width, height))
                    x = int(centerX - (width / 2))
                    y = int(centerY - (height / 2))
                    #print(x,y)
                    boxes.append([x, y, int(width), int(height)])
                    temp.append(LABELS[classID])
                    temp.append(confidence)
                    temp.append(x)
                    temp.append(y)
                    confidences.append(float(confidence))
                    classIDs.append(classID)
                    det.append(temp)
        outlist.append([imgname,(end - start),det])
        show_me(image,classIDs,confidences,boxes,imgname)
    return outlist

def show_me(image,classIDs,confidences,boxes,nam):
    np.random.seed(42)
    COLORS = np.random.randint(0, 255, size=(len(LABELS), 3),
	dtype="uint8")
    idxs = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.3)
    if len(idxs) > 0:
            # loop over the indexes we are keeping
            for i in idxs.flatten():
                    # extract the bounding box coordinates
                    (x, y) = (boxes[i][0], boxes[i][1])
                    (w, h) = (boxes[i][2], boxes[i][3])

                    # draw a bounding box rectangle and label on the image
                    color = [int(c) for c in COLORS[classIDs[i]]]
                    cv2.rectangle(image, (x, y), (x + w, y + h), color, 2)
                    text = "{}: {:.4f}".format(LABELS[classIDs[i]], confidences[i])
                    cv2.putText(image, text, (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX,
                            0.5, color, 2)
    cv2.imwrite(str(nam)+".jpg", image)


                    
initialisation_COCO()
initialisation_Yolo("\yolo416")
print(run_YOLO(import_img("image"+'\\',os.listdir("image"))))
#print(import_img("image"+'\\',["nur.jpg"]))
#print(run_YOLO(import_img("image"+'\\',["ALO.jpg"])))
#print(LABELS)
print(os.listdir("image"))
```
